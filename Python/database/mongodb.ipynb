```python
from pymongo import MongoClient

# 连接到MongoDB
client = MongoClient("mongodb://10.64.3.99:27017/")  # 替换为你的MongoDB连接字符串
db = client["cowave-egg"]  # 替换为你的数据库名称

# 需要搜索和替换的字符串
search_str = "cowave-web.oss-cn-hangzhou.aliyuncs.com"  # 要查找的字符串
replace_str = "cowave-web-1309058320.cos.ap-guangzhou.myqcloud.com"  # 替换的字符串

# 需要处理的集合和字段
collections_to_process = ["News", "NewsEn"]
fields_to_check = ["titleImg", "content", "html"]

# 遍历指定的集合
for collection_name in collections_to_process:
    collection = db[collection_name]

    # 查找包含搜索字符串的文档
    query = {"$or": [{field: {"$regex": search_str}} for field in fields_to_check]}
    documents = collection.find(query)

    for doc in documents:
        updated_doc = {}
        # 仅处理指定的字段
        for field in fields_to_check:
            if (
                field in doc
                and isinstance(doc[field], str)
                and search_str in doc[field]
            ):
                updated_doc[field] = doc[field].replace(search_str, replace_str)
                print(f"已替换字段: {field}, 文档ID: {doc['_id']}\n")

        # 如果有更新的内容才执行更新操作
        if updated_doc:
            collection.update_one({"_id": doc["_id"]}, {"$set": updated_doc})

print("字符串替换完成")
```
