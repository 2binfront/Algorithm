```python
# Definition for singly-linked list.
from typing import List, Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        tmp = []
        for i in lists:
            if i :
                tmp.append(i)
        
        def findSmallest(array):
            idx=None
            min=10000
            for i in range(len(array)):
                if array[i] and array[i].val<min:
                    idx=i
                    min=array[i].val
            return idx
        first=findSmallest(tmp)
        if first is None:
            return None
        head=tmp[first] 
        tail=head
        while tail:
            # print(first)
            tmp[first]=tmp[first].next
            first=findSmallest(tmp)
            if first is None:
                break
            tail.next=tmp[first]
            tail=tail.next
        return head
```
