给你一个链表数组，每个链表都已经按升序排列。

请你将所有链表合并到一个升序链表中，返回合并后的链表。

示例 1：

输入：lists = [[1,4,5],[1,3,4],[2,6]] 输出：[1,1,2,3,4,4,5,6] 解释：链表数组如下： [ 1->4->5, 1->3->4, 2->6 ] 将它们合并到一个有序链表中得到
。 1->1->2->3->4->4->5->6 示例 2：

输入：lists = [] 输出：[] 示例 3：

输入：lists = [[]] 输出：[]

提示：

k == lists.length 0 <= k <= 10^4 0 <= lists[i].length <= 500 -10^4 <= lists[i][j] <= 10^4 lists[i] 按 升序 排列 lists[i].length 的总和不超过
10^4


```python
# Definition for singly-linked list.
from typing import List, Optional


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        tmp = []
        for i in lists:
            if i :
                tmp.append(i)
        
        def findSmallest(array):
            idx=None
            min=10000
            for i in range(len(array)):
                if array[i] and array[i].val<min:
                    idx=i
                    min=array[i].val
            return idx
        first=findSmallest(tmp)
        if first is None:
            return None
        head=tmp[first] 
        tail=head
        while tail:
            # print(first)
            tmp[first]=tmp[first].next
            first=findSmallest(tmp)
            if first is None:
                break
            tail.next=tmp[first]
            tail=tail.next
        return head
```

```python
list1=ListNode(1,ListNode(4,ListNode(5)))
list2=ListNode(1,ListNode(3,ListNode(4)))
list3=ListNode(2,ListNode(6))

print(Solution().mergeKLists([list1,list2,list3]))
```
